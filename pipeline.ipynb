```python
from src.tools import count_all_letters, get_all_frames, fasta_parser, gc, find_stop_codons
```

```python
def fasta_details(file_name):
    """This function performs some general manipulations"""
    d = fasta_parser(file_name)
    for k, v in d.items():
        print(k)
        letter_counts = count_all_letters(v)
        for letter, count in letter_counts.items():
            print(f"{letter} - {count}")
        gc_perc = gc(v)
        print(f"GC% = {round(gc_perc, 2)}")
        frames = get_all_frames(v)
        for f, frame in frames.items():
            stop_codon_indexes = find_stop_codons(frame)
            print(f"Frame: {f}, Stop codons found at: {stop_codon_indexes}")
        print()


fasta_details("data/dna_example_1.fasta")
```
